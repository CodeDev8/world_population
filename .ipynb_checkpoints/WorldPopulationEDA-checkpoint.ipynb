### Exploratory Data Analysis

```python
yearCols = list(df.columns)[5:13][::-1]
```

### <center>Population Contribution by Country</center>

```python
fig, axs = plt.subplots(4, 2, figsize=(18, 18))
fig.suptitle("Population Contribution by Country Through the Decades ", fontsize=18, y=0.95)
paletteColor = sns.color_palette("dark")
for year, ax in zip(yearCols, axs.ravel()):
    top10 = df.groupby("Country")[year].sum().sort_values(ascending=False)[:10]
    ax.set_title(year)
    ax.pie(top10, labels=top10.index, colors=paletteColor, autopct='%.2f')
    ax.axis("equal")
#     top10.plot(kind='pie', top10, labels=top10.index, colors=paletteColor, autopct='%.2f')
plt.show()
```

### <center>Population Change By Area</center>

```python
plt.figure(figsize=(14, 5))
x = df.sort_values("Area (km²)")["Area (km²)"]
y = df.sort_values("Area (km²)")["2022 Population"]
plt.title("Population Change By Area")
plt.xlabel("Area")
plt.ylabel("Population")
plt.plot(x, y)
plt.show()
```

### <center>Population By Continent</center>

```python
fig, axs = plt.subplots(4, 2, figsize=(15,20), sharey=True)
contPop = df.groupby("Continent")[yearCols].sum().reset_index()
plt.subplots_adjust(hspace=0.5)
for year, ax in zip(yearCols, axs.ravel()):
    sns.barplot(ax=ax, data=contPop, x=year, y="Continent", orient="h", palette="viridis")
plt.show()
```

### <center>Human Population Growth</center>

```python
pop = []
year = []
for i in yearCols:
    pop.append(df[i].sum()/1000000000)
    year.append(int(i[:4]))
plt.figure(figsize=(10, 7))
plt.title("Human Population over the Decades")
plt.xlabel("Year")
plt.ylabel("Population in Billion")
sns.lineplot(x=year, y=pop)
plt.show()
```
